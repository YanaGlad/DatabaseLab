## Задание 1
Для каждого артиста вывести все жанры, которые есть в его песнях, и для каждого жанра вывести наиболее продолжительную песню

```python
%%sql
SELECT artist, Milliseconds, track, genre FROM (
  SELECT a.Name artist, t.Milliseconds, 
    row_number() over (partition by a.ArtistId, g.GenreId order by t.Milliseconds desc) trackNum, t.Name track, g.Name genre 
  FROM artists a
  JOIN albums al ON al.ArtistId = a.ArtistId
  JOIN tracks t ON t.AlbumId = al.AlbumId
  JOIN genres g ON g.GenreId = t.GenreId
) WHERE trackNum = 1
```

## Задание 2 
Для каждой компании вывести трек, за который компания заплатила наибольшую сумму, вывести жанр данного трека и его альбом

```python
%%sql
SELECT * FROM (
  SELECT c.CustomerId, c.Company, t.Name track, t.UnitPrice price, g.Name genre, a.Title album,
  row_number() over (partition by c.CustomerId order by ii.UnitPrice * ii.Quantity desc) rn
  FROM customers c
  JOIN invoices i ON i.CustomerId = c.CustomerId
  JOIN invoice_items ii ON ii.InvoiceId = i.InvoiceId
  JOIN tracks t ON ii.TrackId = t.TrackId
  JOIN genres g ON g.GenreId = t.GenreId
  JOIN albums a ON a.AlbumId = t.AlbumId
  WHERE c.Company is not NULL
) WHERE rn == 1
```

## Задание 3
Для каждого плейлиста вывести трек с наибольшей продолжительностью, трек с наименьшей стоимостью (если у треков одинаковая стоимость, то вывести тот, который весит меньше всех), и жанр данного плейлиста (жанром плейлиста называется тот жанр, треков которого больше всех в плейлисте).

```python
%%sql
WITH a as (
  SELECT DISTINCT playlist, first_value(g) over (partition by playlist order by c desc) genre FROM (
    SELECT DISTINCT p.Name playlist, g.Name g, COUNT(*) over (partition by p.PlaylistId, g.GenreId) c
    FROM playlists p
    JOIN playlist_track pt ON pt.PlaylistId = p.PlaylistId
    JOIN tracks t ON t.TrackId = pt.TrackId
    JOIN genres g ON g.GenreId = t.GenreId
  )
)

SELECT DISTINCT p.Name, a.genre,
  first_value(t.Name) over (partition by p.PlaylistId order by t.Milliseconds desc) long_duration,
  first_value(t.Name) over (partition by p.PlaylistId order by t.UnitPrice, t.Bytes) lowest_cost
FROM playlists p
JOIN playlist_track pt ON pt.PlaylistId = p.PlaylistId
JOIN tracks t ON t.TrackId = pt.TrackId
JOIN a ON a.playlist = p.Name
```

## Задание 4

Для каждого жанра вывести треки, которые встречались в большем числе заказов, 
сумму, потраченную, на данные треки, количество купленных треков и компанию, которая больше всех приобрела данную композицию

```python
%%sql
WITH a as (
SELECT DISTINCT TrackId, Name, first_value(Company) over (partition by TrackId order by cnt, Company desc) Cname FROM (
SELECT t.TrackId, t.Name, c.Company,
COUNT(*) over (partition by t.TrackId, c.Company) cnt
FROM tracks t
JOIN invoice_items ii ON ii.TrackId = t.TrackId
JOIN invoices i ON i.InvoiceId = ii.InvoiceId
JOIN customers c ON c.CustomerId = i.CustomerId
WHERE c.Company is not NULL
)
)

SELECT b.genre, b.TrackId, a.Name, b.c, b.price, a.Cname Company
FROM (
SELECT *, row_number() over (partition by genre order by c desc) rn FROM (
SELECT
t.TrackId,
g.Name genre,
SUM(i.Quantity) over (partition by t.TrackId) c,
SUM(i.Quantity * i.UnitPrice) over (partition by t.TrackId) price
FROM genres g
JOIN tracks t ON t.GenreId = g.GenreId
JOIN invoice_items i ON i.TrackId = t.TrackId
)
) b
JOIN a ON a.TrackId = b.TrackId
WHERE rn = 1
```
